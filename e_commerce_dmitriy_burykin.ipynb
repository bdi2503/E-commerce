# Когортный анализ пользователей. 
#### В период с января по декабрь выявляем когорту с самым высоким retention на 3й месяц.
**Для начала определим основные параметры:**

* Признак формирования когорты — действие, которое объединяет людей в группу: дата первого заказа.

* Анализируемый ключевой показатель: retention - сколько процентов пользователей совершили покупку в месяц 0, вернулись и продолжили совершать покупки.

```python
users_orders = ood_buy.merge(olist_customers_datase, how='left', on='customer_id')
users_orders.head()
```

```python
# Извлечем из колонки *order_purchase_timestamp* год и дату создания заказа
users_orders['year_month'] = users_orders.order_approved_at.apply(pd.to_datetime).dt.to_period('M')
```

Определимся с выбором периода для формирования когорт. Для этого изучим, за какой временной интервал в целом мы располагаем данными.

```python
dates = users_orders['year_month'].to_frame().drop_duplicates().dropna().sort_values(by='year_month')
dates
```

Как видно, период с января по декабрь у нас отражен только для 2017 года, его и будем изучать.
* Сначала получим таблицу с датой первого заказа для каждого уникального клиента.

```python
first_purchase = users_orders.groupby(['customer_unique_id', 'order_purchase_timestamp'], as_index=False)\
    .agg({'order_status': 'count'})\
    .sort_values('order_purchase_timestamp', ascending=True)\
    .rename(columns={'order_purchase_timestamp': 'first_order_date'})\
    .drop_duplicates(subset='customer_unique_id', keep='first')\
    .drop('order_status', axis=1)
first_purchase.head()
```

Теперь смержим полученную таблицу с user_orders, преобразуем дату первого заказа в формат "год-месяц" и отсортируем записи с первой датой заказа в пределах 2017 года.

```python
cohort = users_orders.merge(first_purchase, how='left', on='customer_unique_id')
cohort['first_order_date'] = cohort.first_order_date.apply(pd.to_datetime).dt.to_period('M')
cohort = cohort.query('first_order_date > "2016-12" and first_order_date < "2018-01"')
cohort.head()
```

Формируем таблицу для расчета числа уникальных клиентов по месяцам и когортам, добавляем колонку с разницей в месяцах между датами покупок

```python
cohort_table = cohort.groupby(['first_order_date', 'year_month'], as_index=False) \
    .agg({'customer_unique_id': 'nunique'}) 
cohort_table
```

```python
# Создаем когортный период
cohort_table['cohort_period'] = cohort_table.year_month.astype(int) - cohort_table.first_order_date.astype(int)
cohort_table.head()
```

```python
retention_table = cohort_table.pivot_table(index='first_order_date', 
                                           columns='cohort_period', 
                                           values='customer_unique_id')
retention_table
```

Переведем значения в проценты, используя нулевой столбец как делитель.

```python
size = retention_table.iloc[:, 0]
retention_table_percent = retention_table.div(size, axis=0)
retention_table_percent 
```

```python
sns.set(style='white')
plt.figure(figsize=(18, 8))
plt.title('Cohorts: User Retention')
sns.heatmap(retention_table_percent.T, mask=retention_table_percent.T.isnull(), annot=True, fmt='.3%',
            vmin=0, vmax=0.005, cmap="coolwarm");
```

Найдем максимальное значение retention за 3 месяц.

```python
print('Максимальное значение retention за 3 месяц =',
      round(retention_table_percent.iloc[:, 3].max() * 100, 2),
      '% принадлежит к когорте',
      retention_table_percent.iloc[:, 3].idxmax())
```

# RFM-сегментация
Построим RFM-сегментацию пользователей, чтобы качественно оценить аудиторию.

Метрики:
* R - время от последней покупки пользователя до текущей даты
* F - суммарное количество покупок у пользователя за всё время
* M - сумма покупок за всё время

Для каждого RFM-сегмента построим границы метрик recency, frequency и monetary для интерпретации этих кластеров.

Создаем фрейм с уникальным пользователем, заказом и датой покупки

```python
buy = pd.merge(ood_buyM, ocd, how="left", on='customer_id')
buy = buy[['customer_unique_id', 'order_id', 'order_approved_at']].sort_values('order_approved_at')
buy.head()
```

Создаем столбец с общей суммой проданного продукта

```python
product = olist_order_items_dataset.sort_values('order_item_id')
product['sales'] = product.order_item_id * product.price
product = product[['order_id', 'sales']]
```

```python
# Фрейм с общей суммой в заказе
product_sales = product \
    .groupby('order_id', as_index=False) \
    .agg({'sales': 'sum'})
product_sales.head()
```

Фрейм с уникальным пользователем, заказом, датой покупки и общей суммой

```python
buy_product = pd.merge(buy, product_sales, how="inner", on='order_id')
buy_product['order_approved_at'] = buy_product.order_approved_at.dt.date
buy_product['order_approved_at'] = pd.to_datetime(buy_product.order_approved_at)
buy_product.head()
```

```python
# Узнаем самую позднюю дату заказа
max_date = buy_product.order_approved_at.max() + pd.offsets.Day(1)
max_date
```

Создаем столбец, в котором указано количество дней до даты самого позднего заказа

```python
buy_product['last_buy'] = max_date - buy_product.order_approved_at
buy_product['last_buy'] = buy_product['last_buy'].dt.days
buy_product.head()
```

**Создаем rfm фрейм, где:**
* recency - время от последней покупки пользователя до текущей даты,
* frequency - суммарное количество покупок у пользователя за всё время, 
* monetary - сумма покупок за всё время.

```python
rfm_data = buy_product \
    .groupby('customer_unique_id', as_index=False) \
    .agg({'last_buy': 'min',
          'order_id': 'count',
          'sales': 'sum'}) \
    .sort_values('order_id', ascending=False)
rfm_data = rfm_data.rename(columns={'last_buy': 'recency',
                                    'order_id': 'frequency',
                                    'sales': 'monetary'})
rfm_data.head()
```

Попробуем понять, как разбить данные на кластеры. Посмотрим какие значения имеем.

```python
rfm_data.frequency.value_counts()
```

```python
rfm_data.frequency.value_counts().plot()
```

Количество значений распределено крайне неравномерно. Возможно, было бы верно разделить на 3-5 равных групп, но методом quantile сделать это не получится из-за неравномерности распределения.

Выделим следующие группы:
- 1 заказ
- 2-3 заказа
- 4 и более заказов

Для остальных признаков оставим метод quantile для деления на кластеры.

```python
quintiles = rfm_data[['recency', 'monetary']].quantile([.3, .6]).to_dict()
quintiles
```

Напишем функции для определения рангов

```python
# For recency:
def r_score(x):
    if x <= quintiles['recency'][.3]:
        return 3
    elif x <= quintiles['recency'][.6]:
        return 2
    else:
        return 1
    
# For monetary:
def m_score(x):
    if x <= quintiles['monetary'][.3]:
        return 1
    elif x <= quintiles['monetary'][.6]:
        return 2
    else:
        return 3  

# For frequency:
def f_score(x):
    if x == 1:
        return 1
    elif x == 2 or x == 3:
        return 2
    else:
        return 3 
```

На основе функций создаем новые признаки

```python
rfm_data['R'] = rfm_data['recency'].apply(lambda x: r_score(x))
rfm_data['F'] = rfm_data['frequency'].apply(lambda x: f_score(x))
rfm_data['M'] = rfm_data['monetary'].apply(lambda x: m_score(x))

# Создаем rfm_score
rfm_data['rfm_score'] = rfm_data['R'].map(str) + rfm_data['F'].map(str) + rfm_data['M'].map(str)
rfm_data.head()
```

Сгруппируем по новому признаку и посчитаем, сколько клиентов в каждом

Мы получили 23 сегмента. Имеет смысл немного объеденить их в группы. Например, так:
1. leaving:
    - 111, 112 и 113 - leaving_disposable - одноразовые
    - 121, 122 и 123 -leaving_rare - уходящие и редкие
    - 131,  132 и 133 - leaving_permanent - уходящие постоянные
 

2. sleeping:
    - 211, 212 и 213 - sleeping_disposable- разовые 
    - 221,222 и 223 - sleeping_rare - редкие 
    - 233 - sleeping_permanent- постоянные

3. permanent:
    - 311,312 и 313 - permanent_new - новички 
    - 321 и 322 - permanent_rare_econom - постоянные с маленьким чеком
    - 323 - permanent_rare_vip -постоянные с хорошим чеком
    - 333 - VIP - лучшие 

У нас осталось 10 групп. Это уже можно визуализировать.

```python
def segm(x):
    if x in ('111', '112', '113'):
        return 'leaving_disposable'
    elif x in ('121', '122', '123'):
        return 'leaving_rare'
    elif x in ('131', '132', '133'):
        return 'leaving_permanent'
    elif x in ('211', '212', '213'):
        return 'sleeping_disposable'
    elif x in ('221', '222', '223'):
        return 'sleeping_rare'
    elif x == '233':
        return 'sleeping_permanent'
    elif x in ('311', '312', '313'):
        return 'permanent_new'
    elif x in ('321', '322'):
        return 'permanent_rare_econom'
    elif x == '323':
        return 'permanent_rare_vip'
    else:
        return 'VIP'
```

```python
rfm_data['segment'] = rfm_data['rfm_score'].apply(lambda x: segm(x))
rfm_data
```
